```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
#create clustered artificial data
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0]
```

```python
sns.set_style('darkgrid')
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
#choose number of clusters
kmeans = KMeans(n_clusters=4)
```

```python
#fit to features
kmeans.fit(data[0])
```

```python
#returns centers of clusters
kmeans.cluster_centers_
```

```python
#return labels, models thinks to be true for clusters
kmeans.labels_
```

```python
#check how Kmeans compares to actual clusters
#lets create subplot with one row by two columns, share same axes (Sharey=True)
# (ax1,ax2) tuple unpacking to grab these two set of axes
# ax1 plots original data and colors them by Kmeans algorithm
# ax1 plots original data and colors them by actual division

fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax1.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
